```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
C = np.array([2,4])
print(C.ndim)
b = np.arange(0,40,3)
plt.xlim(-5,30)
plt.ylim(-5,30)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(b*C[0], b*C[0], color='green') 
plt.plot(b*C[1], b*C[1], color='k') 
plt.show()
```

$$ C = \left\{\begin{array}\\2x+4y\end{array}\right. $$



```python
C = np.array([
    [5,2,7],
    [4,1,4],
    [9,1,7]
])
fig = plt.figure()
A1 = plt.subplot(projection='3d')
A1.set_xlim([-5, 5])
A1.set_ylim([-5, 5])
A1.set_zlim([-5, 5])
A1.set_xlabel('x-axis')
A1.set_ylabel('y-axis')
A1.set_zlabel('z-axis')
origin = (-5,-5,-5)
A1.quiver(origin, origin, origin, C[:,0], C[:,1], C[:,2], arrow_length_ratio=0.07, colors=['k','orange','violet'])
plt.show()
```

$$ C = \left\{\begin{array}\\ 5x+2y-7z\\ 4x - 1y+4z\\9x - 1y + 7z \end{array}\right. $$
