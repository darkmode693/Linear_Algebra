```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A1 = np.array([
    [-10, 0],
    [0, 10]
])
B = np.arange(0,2,.10)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.scatter(B*A1[0,0],B*A1[1,0], color='red')
plt.scatter(B*A1[0,1],B*A1[1,1], color='blue')
plt.grid()
plt.show()
```

```python
A2 = np.array([
  [2,0],
  [2,2]
])
B2 = np.arange(0,2,.20)
plt.title("The Scale Transformation");
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.scatter(B2*A2[0,0],B2*A2[1,0], color='red')
plt.scatter(B2*A2[0,1],B2*A2[1,1], color='blue')
plt.grid()
plt.show()
```

```python
C1 = np.array([
    [0, 5],
    [6,0]
])
D1 = np.arange(0,2,.10)
plt.xlim(-5,15)
plt.ylim(-5,15)
plt.scatter(D1*C1[0,0],D1*C1[1,0], color='red')
plt.scatter(D1*C1[0,1],D1*C1[1,1], color='blue')
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.grid()
plt.show()
```

```python
C2 = np.array([
    [0, 5],
    [6, 4]
])
D2 = np.arange(0,2,.10)
plt.title('The Shear Transformation');
plt.xlim(-5,15)
plt.ylim(-5,15)
plt.scatter(D2*C2[0,0],D2*C2[1,0], color='red')
plt.scatter(D2*C2[0,1],D2*C2[1,1], color='blue')
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.grid()
plt.show()
```
