```python
V1 = np.array ([1,3,1])
V2 = np.array ([1,4,5])
G = np.arange(-13,13,.5)
a1,a2 = np.meshgrid(G,G)
V = V1+ V2
sX = a1*V1[0] + a2*V2[0] 
sY = a1*V1[1] + a2*V2[1] 
plt.scatter(sX,sY, s=2, alpha=1)

plt.axhline(y=0, color='red')
plt.axvline(x=0, color='red')
plt.grid()
plt.show()

```
